```python
class Humano:
    def __init__(self,edad,nombre):
        self.edad = edad
        self.nombre = nombre
    def mostrarDatos(self):
        print("Me llamo",self.nombre,"y mi edad es",self.edad)
```

**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x)
plt.plot(x,y,label = "sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(-10,10,100)
m = 2
b = 1
y = m*x + b
plt.plot(x,y,label = f'Recta: y = {m}x + {b}')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```
